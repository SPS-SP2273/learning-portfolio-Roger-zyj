## Exercise 1 (Celcius to Farenheit)

```python
temperatures_celcius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

def celcius_to_farenheit(cel):
    return (cel * 9/5) + 32

for cel in temperatures_celcius:
    print(f"{cel} degrees Celcius == {celcius_to_farenheit(cel)} degrees Farenheit")
```

## Exercise 2 (Multiplication table)

```python
for i in range(1, 5+1):
    print(f"{i} : ", end="\t")
    for j in range(1, 5+1):
        print(f'{i*j}', end='\t')
    print('\n')
```
