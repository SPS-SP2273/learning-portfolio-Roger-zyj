## Exercise 1 (Make me an odd list)

```python
odd_squares = []
for i in range(10):
    if i%2 == 0:
        continue
    odd_squares.append(i**2)
print(odd_squares)
```

## Exercise 2 (Make me another odd list)

```python
odd_squares_comprehension = [i**2 for i in range(1, 10, 2)]
odd_squares_comprehension = [i**2 for i in range(10) if i%2==1]
print(odd_squares_comprehension)
```

## Exercise 3 (Time me!)

```python
%%timeit
odd_squares = []
for i in range(10):
    if i%2 == 0:
        continue
    odd_squares.append(i**2)
```

```python
%%timeit
odd_squares_comprehension = [i**2 for i in range(10) if i%2==1]
```

## Exercise 4 (A problem of decay)

```python
current = initial = 100
years_passed = 0

while current >= 0.5*initial:
    current *= 0.95
    years_passed += 1

print(years_passed)
```

## Exercise 5 (Changes in CO<sub>2</sub>) 

```python
import numpy as np
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

arr_curr = np.array(co2_data)
arr_prev = np.array( [(1999, 369.55)] + co2_data[0:-1] )
Delta_ppm = arr_curr[:, 1] - arr_prev[:, 1]
print(arr_curr[Delta_ppm >= 3][0][0] , Delta_ppm[Delta_ppm >= 3][0])
```
