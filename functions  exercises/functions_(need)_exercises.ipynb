## Exercise 1 (Do you know why?)

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

We can think about it this way:

By default, the outer-most "return" keyword will always be executed. This means that by default, the function will return "Hello {name}!".
However, in the special case where name=="Batman", the "return" keyword in the "if" block will be executed. And whenever a function executes a "return" keyword, the rest of the function will not be executed, so the outer-most "return" keyword will not be reached. Thus, the function will return "Hello Batman! So, nice to meet you!" without returning "Hello Batman!".

## Exercise 2 (Chubby or not)

```python
def calculate_bmi(weight_kg, height_m):
    categories = {
        "Underweight":      (-float("inf"), 18.5),
        "Normal weight":    (18.5, 25),
        "Overweight":       (25, 30),
        "Obesity":          (30, float("inf"))
    }

    BMI = weight_kg / (height_m ** 2)
    print(BMI)
    for category in categories:
        lowerBound, upperBound = categories[category]
        if lowerBound <= BMI < upperBound:
            return category
        
print(calculate_bmi(56, 1.75))
# print(calculate_bmi(63.9, 1.6))
```
