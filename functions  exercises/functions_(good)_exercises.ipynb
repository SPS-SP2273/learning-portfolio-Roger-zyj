## Exercise 1 (Celsius to Fahrenheit or Kelvin)

```python
def convert_celsius(temperature_celsius, target_scale="Fahrenheit"):
    if target_scale == "Fahrenheit":
        return (temperature_celsius * 9/5) + 32
    elif target_scale == "Kelvin":
        return temperature_celsius + 273.15
    else:
        print("?")

print(convert_celsius(25))
print(convert_celsius(25, target_scale="Kelvin"))
```

## Exercise 2 (Fahrenheit to Celsius or Kelvin)

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale="Celsius"):
    if target_scale == "Celsius":
        return (temperature_fahrenheit - 32) * 5/9
    elif target_scale == "Kelvin":
        return (temperature_fahrenheit - 32) * 5/9 + 273.15
    else:
        print("?")

print(convert_fahrenheit(32))
print(convert_fahrenheit(32, "Kelvin"))
```

## Exercise 3 (General Temperature Conversion)

```python
def convert_temperature(temperature, source_scale, target_scale):
    if source_scale == "Celsius":
        return convert_celsius(temperature, target_scale=target_scale)
    elif source_scale == "Fahrenheit":
        return convert_fahrenheit(temperature, target_scale=target_scale)
    elif source_scale == "Kelvin":
        if target_scale == "Celsius":
            return temperature - 273.15
        elif target_scale == "Fahrenheit":
            return (temperature - 273.15) * 9/5 + 32
    else:
        print("?")

print(convert_temperature(0, "Kelvin", "Celsius"))
```
