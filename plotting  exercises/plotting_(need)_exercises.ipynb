### Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color="brown", linestyle="--",
             marker="o", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="X vs Y")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(["Y values with errors"], loc="lower right")
plt.show()
```

### Exercise 2 (A simple mathematical plot) 

```python
import numpy as np

# 1. Using numpy to generate x and y data
x = np.linspace(0, 50, 1000)
y = np.exp(-x/10) * np.sin(x)

# Customising the plot.
# 1. The x and y axes have labels of fontsize of 15.
# 2. The x and y axes labels are in the color, darkorange.
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange', rotation=0)

# 3. There is a grid with an opacity of 25%
plt.grid(alpha=0.25)

# 4. Indicate the equation of the function in the title with a font size of 20.
plt.title('$y = e^{-\\frac{x}{10}} sin(x)$', fontsize=20)

# 5. Change the  limit (using ylim) to [-1, 1]
plt.ylim(-1, 1)

plt.plot(x, y)
```
