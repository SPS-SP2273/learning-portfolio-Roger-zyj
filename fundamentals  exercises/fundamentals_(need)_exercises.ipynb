## Exercise 4 (More than one way to divide)

```python
5/2         # This is the standard division. Takes in 2 numerical values and returns a float.
```

```python
5//2        # This is the floor division function. Takes in 2 numerical values and returns the largest integer that is smaller than the quotient.
```

```python
5%2         # This is the remainder function. Takes in A%B and returns the remainder when A is divided by B.
```
