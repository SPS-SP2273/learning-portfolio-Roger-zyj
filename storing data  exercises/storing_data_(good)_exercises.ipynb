## Exercise 2 (Show me the "odd" letters)

```python
import numpy as np
np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])

print( np_array_2d[:, 1][(np_array_2d[:, 0].astype("int") % 2) == 1] )
```
