## Exercise 1 (Tidying a collaboration) 

### Task 1 and 2

```python
def rename_my_file(old_file_name):
    char_to_num = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    old_file_name = old_file_name.replace("_", "-")
    l = old_file_name.split("-")
    return char_to_num.get(l[0], l[0]) + "-" + "-".join(l[1:])

print(rename_my_file("may-10-singapore.txt"))
print(rename_my_file('oct-08_singapore.txt'))
print(rename_my_file("10-08-singapore.txt"))
```

### Task 3

```python
old_file_names = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

new_file_names = map(rename_my_file, old_file_names)
for old, new in list(zip(old_file_names, new_file_names)):
    print(f"{old} ----> {new}")
```

### Task 4

```python
import os
import glob

# print(os.path.join("os-collaboration-exercise-data", "*.txt"))
file_names = glob.glob(os.path.join("os-collaboration-exercise-data", "*.txt"))

for old_file_name in file_names:
    os.rename(old_file_name, rename_my_file(old_file_name))
```

### Task 5

```python
city_list = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in city_list:
    try:
        os.mkdir(os.path.join("os-collaboration-exercise-data", city))
    except FileExistsError:
        print(f"os-collaboration-exercise-data/{city} alr exist")
```

### Task 6

```python
import shutil

Cairo_file_names = glob.glob(os.path.join("os-collaboration-exercise-data", "*cairo*"))
for file_name in Cairo_file_names:
    shutil.copy(file_name, "os-collaboration-exercise-data/Cairo")
```

### Task 7

```python
for city in city_list:
    city_file_names = glob.glob(os.path.join("os-collaboration-exercise-data", f"*{city}*".lower()))
    for file_name in city_file_names:
        new_path = os.path.join("os-collaboration-exercise-data", city, str(rename_my_file(file_name[len("os-collaboration-exercise-data")+1:])) )
        shutil.move(file_name, new_path)
```
